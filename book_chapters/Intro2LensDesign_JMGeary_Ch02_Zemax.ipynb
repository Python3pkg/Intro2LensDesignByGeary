##### Odds and Ends

**More on f-number**

In the following formulae, 

* $U'$ is the "bend angle" of the ray coming to focus in the image space (refer to figure 2.7 in the book)
* $EFL$ is the Effective Focal Length
* $EPD$ is the Entrance Pupil Diameter

image space f-number: $f/\# = \frac{\text{EFL}}{\text{EPD}}$

paraxial working f-number: $f/\# = \frac{1}{2 tan (U')}$

working f-number: $f/\# = \frac{1}{2 sin (U')}$

* The image space f-number is equal to the paraxial working f-number when the object is at infinity.
* For finite-conjugates the f-number is called "effective f-number"
* The working f-number is applicable to real aberrated systems where $U'$ departs from its ideal unaberrated path.
