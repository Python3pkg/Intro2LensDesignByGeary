# Chapter 05: Stops and Pupils

The stop is the *limiting* aperture of the sytem, which allows some rays to pass through to the image while blocking others.


Stops define two very important parameters of lenses:
* marginal ray
* chief ray

In fact, these two rays are enough to completely define a paraxial optical system.

## Entrance and Exit pupils

**Entrance pupil:**

The image of the limiting aperture formed by the optical elements that lie towards the object space is the entrance pupil. In other words, it is the entrance pupil is the pupil as seen from the object space. It is usually specified in the lens units.


The relationship between the system F/#, the effective focal length (EFL) and the entrance pupil diameter (EPD) is:

$$
F/\# = \frac{EFL}{EPD}
$$


**Exit pupil:**

The image of the limiting aperture formed by the optical elements that lie towards the image space is the exit pupil.


The stop, entrance pupil and exit pupil are images of one another.

The books shows many different configurations and the corresponding positions of the entrance and the exit pupils. Please refer to the book for details.

In addition, the following references are useful:

```python
# Utility functions to insert surfaces to represent the location
# and diameter of the Entrance and Exit pupils in an optical system
# Notes:
# 1. Unfortunately the complete process couldn't be automated, due
#    to the inability to control the rendering of rays on surfaces
#    After executing these functions, switch over to the Zemax main
#    application, and select "Skip Rays To This Surface" under the 
#    surface properties for the surfaces marked as "Ent. pupil" and
#    "Ext. pupil".
# 2. For cases of very wide-angle lenses, the rays entering the 
#    front surface may look un-realistic sometimes, although the 
#    position and size of the two pupils are just fine.

def zInsertEntranceAndExitPupils(ln):
    """Insert a pair of surfaces in the LDE to represent the 
    entrance and exit pupil positions and diameters.
    
    Parameters
    ----------
    ln : pyzdde link object
    """
    zInsertEntrancePupil(ln)
    zInsertExitPupil(ln)

def zInsertEntrancePupil(ln):
    """Insert a surface in the LDE to represent the entrance 
    pupil position and diameter
    
    Parameters
    ----------
    ln: pyzdde link object
    """
    pupil = ln.zGetPupil()
    ln.zInsertSurface(1)
    ln.zSetSurfaceData(surfaceNumber=1, code=1, value='dummy')
    ln.zSetSurfaceData(surfaceNumber=1, code=3, value=pupil.ENPP)
    # make the semi-diameter of this surface fixed; else 
    # zemax will automatically change it
    ln.zSetSolve(1, 3, 1, 0) # 3=semi-diam, 1=Fixed, 0=Value (ignored)
    ln.zSetSurfaceData(surfaceNumber=1, code=5, value=0)
    ln.zInsertSurface(2) 
    ln.zSetSurfaceData(surfaceNumber=2, code=1, value='Ent. pupil')
    ln.zSetSurfaceData(surfaceNumber=2, code=3, value=-pupil.ENPP)
    ln.zSetSolve(2, 3, 1, 0)
    ln.zSetSurfaceData(surfaceNumber=2, code=5, value=pupil.ENPD/2.0)

def zInsertExitPupil(ln):
    """Insert a surface in the LDE to represent the exit pupil 
    position and diameter

    Parameters
    ----------
    ln: pyzdde link object
    """
    imaSurf = ln.zGetNumSurf()
    pupil = ln.zGetPupil()
    ln.zInsertSurface(imaSurf)
    ln.zSetSurfaceData(imaSurf, code=1, value='dummy')
    ln.zSetSurfaceData(imaSurf, code=3, value=pupil.EXPP)
    # make the semi-diameter of this surface fixed; else 
    # zemax will automatically change it
    ln.zSetSolve(imaSurf, 3, 1, 1) # 3=semi-diam, 1=Fixed, 0=Value (ignored)
    ln.zSetSurfaceData(imaSurf, code=5, value=0)
    ln.zInsertSurface(imaSurf+1) 
    ln.zSetSurfaceData(imaSurf+1, code=1, value='Ext. pupil')
    ln.zSetSurfaceData(imaSurf+1, code=3, value=-pupil.EXPP)
    ln.zSetSolve(imaSurf+1, 3, 1, 0)
    ln.zSetSurfaceData(imaSurf+1, code=5, value=pupil.EXPD/2.0)
```

## The Lagrange Invariant

Excerpt from Zemax training material on optical design:

>The Lagrange invariant equates the marginal and chief ray height-angle products together



$$
n'(\bar{u}'y - u'\bar{y}) = n(\bar{u}y - u\bar{y})
$$

where,

* $u$ and $u'$ are the ray angles for the marginal ray in the left and right side of the plane respectively, 
* $\bar{u}$ and $\bar{u}'$ are the ray angles for the chief ray in the left and right side respectively,
* $y$ is the height of the marginal ray at the plane
* $\bar{y}$ is the height of the chief ray at the plane
* $n$ and $n'$ are the indices in the left and right side of the plane respectively
