## Ray fans and OPD fans

**Ray fan plots**


The ray fan shows the sagittal (x) and/or the tangential (y) components of the *transverse ray aberrations*<sup>*</sup> ($e_x$ and $e_y$ respectively) as a function of the pupil coordinates (Px or Py). The points on the ray fan plot indicate the difference between the ray intercept coordinate and the chief ray intercept coordinate [[1](#r1)].

<sup>*</sup> Note that the ray aberrations include both transverse ray aberrations and longitudinal ray aberrations.  


**OPD fan plots**

The optical path difference (OPD) for a given ray is the difference between the optical path length (OPL) of the ray and the OPL of the chief ray [[1](#r1)]. This difference can be referenced back to the difference between the ray path lengths at the system exit pupil.


** How are they related?**

The relationship between ray aberrations and wavefront aberrations is given by:

$$
e_x = - \frac{R}{n r}\frac{\partial W}{\partial x}
$$

$$
e_y = - \frac{R}{n r}\frac{\partial W}{\partial y}
$$

where, $R$ is the reference sphere radius, $n$ is the refractive index, and $r$ is the radius of the pupil. The reference sphere radius is equal to the distance of the exit pupil from the paraxial image plane for (and containing) a particular field point.

Note that the fan plots are not a complete description of the aberrations in a system as they are only "slices". Although, they are good diagonastic plots for qualitative description of the nature of aberrations in a rotationally symmetric optical system, they may not be sufficient to describe aberrations in non-rotationally symmetric optical systems.

In order to parse the files we will write a small function (not that generic, and not that great ... but that's not the point here)

```python
import re

def readRayFanData(filename, numRays, which='t'):
    """return the fan data points
    
    filename : full file name of the fan data file
    numRays : number of rays used in the settings of Ray/OPD fan plot
    which : 't' = 'Tangential fan'; else 'Sagittal fan'

    The number of data points returned = 2*numRays + 1The number of data points returned = 2*numRays + 1
    """
    x = []
    n = []
    if which == 't':
        sectionHeader = 'Tangential fan'
    else:
        sectionHeader = 'Sagittal fan'
    linesToStore = 2*numRays + 1
    dataSection = False
    startReading = False
    lineCount = 0
    # identify the first occurrence of pattern such as
    # " -1.000      -1.353850"
    secIdPattern = re.compile(sectionHeader)
    dataPattern = re.compile(r'\s*(-?\d?\.\d{3,8}\s*){2}')
    for line in open(filename, 'r'):
        if re.match(secIdPattern, line.rstrip()):
            dataSection = True
        if dataSection and re.match(dataPattern, line.strip()):
            startReading = True
        if startReading and (lineCount < linesToStore):
            n.append(float(line.split()[0]))
            x.append(float(line.split()[1]))
            lineCount += 1
    return n, x
```

```python
def createCenterSpines(c='k'):
    """Create spines at the center"""
    ax = plt.gca() 
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['bottom'].set_color(c)
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    ax.spines['left'].set_color(c)
    ax.tick_params(color=c,labelcolor=c)
    ax.xaxis.set_ticks_position('bottom')  
    ax.yaxis.set_ticks_position('left') 

def decoratePlot():
    ax = plt.gca()
    createCenterSpines(c='#505050')
    ax.grid(color='k')
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(n_ray, y_ray, 'b', n_opd, y_opd, 'r')
decoratePlot()
ax.legend(['Ray Fan', 'OPD Fan'], frameon=False)
ax.set_xlabel(r' $e_y, \, w$', labelpad=-180, fontsize=20, 
              ha='left')
ax.set_ylabel(r'$P_y$', labelpad=-235, fontsize=18, 
              va='bottom', rotation='horizontal')
plt.show()
```

We will convert the data, which are currently Python lists, into Numpy arrays so that we can apply element based operations easily.

```python
import numpy as np
n_ray = np.array(n_ray)
n_opd = np.array(n_opd)
y_ray = np.array(y_ray)
y_opd = np.array(y_opd)
```

```python
def firstDifference(x, dn):
    """compute numerical derivative of a sequence x(n)
    """
    y = (x[1:] - x[0:-1])/dn
    return y
```

```python
# example of firstDifference. numerical derivative of sin
n = np.linspace(-2*np.pi, 2*np.pi, 201)
dn = 4*np.pi/200
x = np.sin(n)
y = firstDifference(x, dn)

plt.plot(n[1:], x[1:], n[1:], y, 'm')
decoratePlot()
plt.legend(['sin', 'first diff'], frameon=False, loc='lower left')
plt.show()
```

Take the derivative of the OPD curve

```python
delta_n = np.abs(n_opd[1] - n_opd[0])
y_opd_dash = firstDifference(y_opd, delta_n)
```

```python
plt.plot(n_opd[1:], y_opd[1:], n_opd[1:], y_opd_dash, 'm')
decoratePlot()
plt.legend(['OPD fan', 'first diff'], frameon=False, loc='lower left')
plt.show()
```

```python
radiusOfCurv = 108.0596
#refIndex = 1.5168
#wavelength = 0.4861
exitPupilDiam = 36.25844

#scaleFactor = - radiusOfCurv/exitPupilDiam

scaleFactor = - wavelength*radiusOfCurv/(0.5*exitPupilDiam)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.plot(n_ray[1:], y_ray[1:], 'b', n_opd[1:], scalefactor*y_opd_dash, 'r--')
decoratePlot()
ax.legend(["Ray Fan", "OPD Fan' (scaled)"], frameon=False,
          loc='lower right')
plt.show()
```
