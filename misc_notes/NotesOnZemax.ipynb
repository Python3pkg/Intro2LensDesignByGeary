### Use of Snell's Law in Zemax

* Zemax always use the full Snell's law i.e. $n_1 sin \theta_1 = n_2 sin \theta_2$ for real ray-tracing (and not the linear approximation) [[1](#r1)]. In fact, it uses the vector form of the Snell's law: $n_1 (N \times k_1) = n (N \times k)$ where $N$ is the unit normal vector of the surface at the point of intersection, and $k$ is the ray direction cosine vector [[2](#r2)].
* Zemax uses the linear approximation ($n_1 \theta_1 = n_2 \theta_2$) to calculate paraxial quantities such as focal length, pupil plane locations and sizes, system magnification, etc. These are also known as first-order or paraxial approximations.
* Zemax uses the third order approximations to characterize common aberrations.

### Paraxial (first order) approximations in Zemax

Zemax uses first order approximation (paraxial) to calculate the following parameters :

* Optical power
* Image locations
* Magnifications
* Effective focal length
* Pupil locations and sizes
* Aperture and F/#

### Typical approach for a simple singlet design in Zemax

Given the optical specification, these are the typical steps in Zemax [[1](#r1), [3](#r3)]:


1. Set the system aperture
   * This is typically in the form of the EPD, or Image space F/# for infinite conjugate systems. Other options are Object Space NA and Float by stop.
   * The lens units (if it is different from the default settings)
2. Set the fields. Options are:
    * angle: field angles are measured w.r.t. the center of the entrance pupil. Positive field angles imply positive slope for the ray in the direction of propagation, and thus refer to negative object coordinates. This option is most useful when at infinite conjugates.
    * object height: cannot be used at infinite conjugates
    * paraxial image height: works well only for systems which are well described by paraxial optics.
	* real image height:  ray tracing with this option is slightly slower since Zemax must use an iterative approach to determine the proper real ray coordinates of the chief ray on the IMA plane.
    * also see the [comment about field specifications](#Comment-about-field-specification-for-rotationally-symmetric-systems) 
3. Set the wavelength
4. Insert surfaces in the LDE
5. Define the glass type
6. Define starting point variables. Note that ALL of these variables are just starting points, and will typically be used as degrees of freedom (optimization variables) during optimization. 
    * Choose an appropriate initial thickness for the glass (typically between 4 mm and 12 mm). 
    * Set the thickness of the last surface equal to EFL
    * Use the thin lens power equation to find the radius of curvature for an equiconvex lens. $R = 2(n-1)f$. Set the radius of the first glass surface to be equal to $R$.
    * We can leave the thickness between a stop surface (if there is any) and the glass surface equal to zero or some positive distance.
7. Define a solve to control Effective Focal Length (EFL) or F/#
    * One of the most common solve to use is a F/# solve (F) on the radius-of-curvature of the lens element's second surface. The F Number solve adjusts the final optical surface curvature to maintain the system focal length. This will allow the initial design to achieve the paraxial EFL (since the object is at infinity) and F/#. Also, Zemax will maintain this F/# during optimization according to the relationship: $\theta = -\frac{1}{2*F/\#}$ 
8. Use Quick focus to focus (this will change the thickness of the last surface before IMA).
9. Determine the degrees of freedom. This is generally the same as the variables initially set in step 6. 
10. Define a Merit function. Enforce constraints on the merit function to eliminate solutions that we do not want. e.g. thickness between air and glass surfaces.
11. Define the optimization variables (degrees of freedom)

The following figure shows how to enforce constraints on the merit function. This is important to set. If this is not set, the Zemax optimization can result in very large thickness between the surfaces. This is because the power of a surface is given as:

$$
\phi = \phi_1 + \phi_2 - \phi_1 \phi_2\frac{t}{n}
$$

and since $\phi_1 < 1$ and $\phi_2 < 1$, $\phi_1 \phi_2 << 1$. Therefore, the thickness must be very large to have any significant effect on the merit function.

### Solves or Optimization variables

When given constraints on an optical design, there are two possible methods of upholding these constraints:
	
* Make the parameters which affect these constraints variables ("V") and add boundary constraints into the Merit Function Editor (to be introduced shortly) OR

* Use special solves to enforce the constraints, eliminating unnecessary variables.

The latter of these two is far superior.Though both provide a method of adjusting lens parameters to maintain a specified constraint, boundary constraints can slow down execution of the merit function.

The optimization time is proportional to (# operands)^2.

### Comment about field specification for rotationally symmetric systems

If the optical system being designed is radially symmetric, it is recommended that one should only specify the field along +Y axis, in which case Zemax assumes that the system is rotationally symmetric and applies symmetry rules for efficient computations. If one specifies both X and Y field coordinates, even if the system is rotationally symmetric, Zemax doesn't apply symmetry based optimization, and it computes along all directions. This will cause the computations to take much longer to finish [[1](#r1)].

### Rules for using Coordinate Breaks manually

1. Global vertex report: Before starting on CBs, open the prescriptio file and make/keep a note of the position/ vertices and orientation of all surfaces (especially the image surface).
2. Pairs of CB: Almost all (with some exceptions) coordinate breaks are done in pairs -- first one for decentering and tilting and the second one for restoring the original axis. Note that the surfaces after the second CB, should retain their vertex positions and orientations as it was before CB was applied. 
3. Pick-up solves: It is recommended to use pick-up solves on the tilt and decenter parameters of the second CB by setting -1 scale factor to lock the second CB to the first. 
4. Co-locate the two coordinate breaks: 
    - Use a dummy surface just after the second CB. 
    - Use a position solve (on the thickness of the surface before the second CB) to track back through all the surfaces.
    - Use a pickup solve (on the thickness of the second CB) to restore the second CB's position at the back of the lens. 
5. Order flag for compound tilts and decenters: If there are multiple tilts or mixture of tilt and decenter, set the order flag on the second CB to 1. This is important, as it will execute second CB in reverse order.
6. General recommendation: Unless there is a special case or need (such as using single CB), use the Tilt/Decenter Element Tool provided by Zemax.
